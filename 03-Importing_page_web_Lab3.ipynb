# parse the web page and get the header row of the table

```python
# Récupérer le contenu de la page web
# Ouvrir le fichier local
with open("mypage.html", "r", encoding="utf-8") as file:
    content = file.read()
bs = BeautifulSoup(content, "lxml")
theadrows = bs.find('table', {'id':'tblDeaths'}).thead.find_all('th')
```

```python
print(theadrows)
```

```python
type(theadrows)
```

```python
labelcols = [j.get_text() for j in theadrows]
print(labelcols)
labelcols[0] = "rowheadings"
print(labelcols)
```

# Get the data from the table cells

```python
rows = bs.find('table', {'id':'tblDeaths'}).tbody.find_all('tr')
datarows = []
labelrows = []
for row in rows:
  rowlabels = row.find('th').get_text()
  cells = row.find_all('td', {'class':'data'})
  if (len(rowlabels)>3):
    labelrows.append(rowlabels) 
  if (len(cells)>0):
    cellvalues = [j.get_text() for j in cells]
    datarows.append(cellvalues)
```

# Insérer au début de chaque ligne de datarows le nom du pays qui sont stokés dans labelsrows en utilisant la fonction insert

```python
for i in range(len(datarows)):
  datarows[i].insert(0, labelrows[i])

print(datarows[0:4])
print(datarows)
```

# Load the data into pandas

```python
totaldeaths = pd.DataFrame(datarows, columns=labelcols)
```

# Correction des noms des colonnes et conversion des données en valeurs numériques

```python
totaldeaths.columns = totaldeaths.columns.str.replace(" ", "_").str.lower()
```

```python
totaldeaths
```

```python
for col in totaldeaths.columns[1:-1]:
  totaldeaths[col] = totaldeaths[col].str.replace("[^0-9]","").astype('int64')
```
